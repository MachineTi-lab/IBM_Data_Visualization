```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import folium
%matplotlib inline
```

```python
import requests
import io
```

```python
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Historical_Wildfires.csv"
response = requests.get(URL)

text = io.BytesIO(response.content)

df = pd.read_csv(text)
```

```python
df.head()
```

```python
df.columns 
```

```python
df.dtypes
```

```python
import datetime as dt

df["Year"] = pd.to_datetime(df["Date"]).dt.year
df["Month"] = pd.to_datetime(df["Date"]).dt.month 
```

```python
df.head(5)
```

```python
df.dtypes
```

```python
plt.figure(figsize =(12,6))
df_new = df.groupby("Year")["Estimated_fire_area"].mean()

df_new.plot(x = df_new.index , y= df_new.values)
plt.xlabel("Year")
plt.ylabel("Avarage Estimated Fire Area")
plt.show()
### there is a peak between 2010 and 2013.
```

```python
plt.figure(figsize= (12,6))

df_ym = df.groupby(["Year","Month"])["Estimated_fire_area"].mean()

df_ym.plot(x = df_ym.index , y = df_ym.values)
plt.xlabel('Year, Month')
plt.ylabel('Average Estimated Fire Area (km)')
plt.title('Estimated Fire Area over Time')
plt.show()
```

```python
df["Region"].unique()
```

```python
plt.figure(figsize = (12,6)) 

df_re = df.groupby(["Region"])["Estimated_fire_area"].mean()

sns.barplot(data = df , x = df["Region"] , y = df["Estimated_fire_area"]) 
plt.xlabel("Region")
plt.ylabel("Estimate fire area")
plt.title('Distribution of Mean Estimated Fire Brightness across Regions')
plt.show()
```

```python
region_counts = df.groupby(["Region"])["Count"].sum()
region_counts 
```

```python
plt.pie(region_counts, labels=region_counts.index)
plt.title('Percentage of Pixels for Presumed Vegetation Fires by Region')
plt.axis('equal')
plt.legend()
[(i,round(k/region_counts.sum()*100,2)) for i,k in zip(region_counts.index, region_counts)]
           
plt.show()
```
